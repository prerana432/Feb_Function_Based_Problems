```python
def analyze_engagement(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

analyze_engagement([200, 300, 550])
```

```python
def check_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

check_stock(6)
```

```python
def rainfall_check(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_check([70, 75, 65, 78], 70)
```

```python
def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

# Example
check_duplicates(["user1", "user2", "user3", "user1"])
```

```python
def check_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

check_eligibility(21)
```

```python
def premium_crops(prices):
    premium_list = []

    for price in prices:
        if price > 2000:
            premium_list.append(price)

    print("Premium Crops:", premium_list)

premium_crops([1500, 2500, 1800, 3200])
```

```python
def system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

system_health(7)
```

```python
def transaction_check(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_check(60000)
```

```python
def attendance_check(attended_classes, total_classes):
    percentage = (attended_classes / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_check(40, 50)
```

```python
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Units:", units)
    print("Total Bill:", bill)
    print("Usage Level:", usage)

electricity_bill(250)
```
